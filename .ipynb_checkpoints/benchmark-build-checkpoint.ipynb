```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import pickle

# This uses the Answers.csv file from the 10% Stack Overflow data
answer_file = "data/Answers.csv"
# The output is an edge list file, where each row represents a user-question interaction, i.e. user A has answered question b
# that the algorithm should predict
benchmark_1_data_file = "output/benchmark_1.txt"
benchmark_2_data_file = "output/benchmark_2.txt"
benchmark_3_data_file = "output/benchmark_3.txt"

# intermediate file to reduce accelerate analysis
simplified_data_file = "processed_data/simplified_df.pkl"

# Question_ids and user_ids may overlap, but that does not mean questions are users!!!
# Diff Soln: append a header in ids. This may help us keep track of questions and users
question_header = 'q_'
user_header = 'u_'
# tag_header = 't_'
```

## Pre-processing

```python
df = pd.read_csv(answer_file, encoding="ISO-8859-1")
df.head(5)
```

```python
# simplify our analysis to only users and questions
simplified_df = df[['OwnerUserId', 'ParentId']]
simplified_df = simplified_df.dropna()
simplified_df = simplified_df.assign(question_id=lambda x: [question_header + str(i) for i in x.ParentId])
simplified_df = simplified_df.assign(user_id=lambda x: [user_header + str(int(i))for i in x.OwnerUserId])

# 2.a) just a small sanity check
tmp_a = [(i, j ) for i, j in zip(simplified_df['ParentId'], simplified_df['question_id']) if question_header + str(i) != j]
tmp_b = [(i, j ) for i, j in zip(simplified_df['OwnerUserId'], simplified_df['user_id']) if user_header + str(int(i)) != j]
assert len(tmp_a) == 0 and len(tmp_b) == 0

# 2.b) drop the old columns
simplified_df = simplified_df.drop(['ParentId'], axis=1)
simplified_df = simplified_df.drop(['OwnerUserId'], axis=1)
simplified_df.head(5)
```

```python
# save this to disk
simplified_df.to_pickle(simplified_data_file)
```

```python
# now load it back...
simplified_df = pd.read_pickle(simplified_data_file)
simplified_df.head(5)
```

## Direction I: users who have answered many questions
Intuition: the more questions a user answers, the more we know about the user

```python
user_question_df = simplified_df.groupby(['user_id'])['question_id'].apply(list).reset_index(name='question_list')
user_question_df["num_question"] = user_question_df["question_list"].apply(lambda x: len(x))
user_question_df.head(5)
```

```python
user_question_df.shape
```

```python
all_num_question = list(user_question_df['num_question'])
plt.xlim(0, 2000)
plt.ylim(0, 500)
plt.xlabel('number of users')
plt.ylabel('number of questions a user has answered')
plt.title('distribution of number of questions that each user has answered')
plt.plot(sorted(all_num_question, reverse=True))
```

```python
num_question_per_user_thre = 50
common_users = set(user_question_df[user_question_df.num_question > num_question_per_user_thre]['user_id'])
print(len(common_users))
benchmark_1 = simplified_df[simplified_df['user_id'].isin(common_users)]
print(benchmark_1.count())
random_subset_b1 = benchmark_1.sample(n=3000)
```

```python
random_subset_b1.to_csv(benchmark_1_data_file, sep=' ', header=False, index=False)
```

## Direction II: questions that have been answered by many users
Intuition: a popular question that can always appreciate some new input

```python
question_user_df = simplified_df.groupby(['question_id'])['user_id'].apply(list).reset_index(name='user_list')
question_user_df["num_user"] = question_user_df["user_list"].apply(lambda x: len(x))
question_user_df.head(5)
```

```python
question_user_df.shape
```

```python
all_num_user = list(question_user_df['num_user'])
plt.xlim(0, 2000)
plt.ylim(0, 100)
plt.xlabel('number of questions')
plt.ylabel('number of users a question has')
plt.title('distribution of number of users that each question has')
plt.plot(sorted(all_num_user, reverse=True))
```

```python
num_user_per_question_thre = 10
popular_questions = set(question_user_df[question_user_df.num_user > num_user_per_question_thre]['question_id'])
print(len(popular_questions))
benchmark_2 = simplified_df[simplified_df['question_id'].isin(popular_questions)]
print(benchmark_2.count())
random_subset_b2 = benchmark_2.sample(n=3000)
```

```python
random_subset_b2.to_csv(benchmark_2_data_file, sep=' ', header=False, index=False)
```

## Direction III: question-user pairs for popular questions and valuable users

```python
num_question_per_user_thre = 50
num_user_per_question_thre = 15
popular_questions = set(question_user_df[question_user_df.num_user > num_user_per_question_thre]['question_id'])
common_users = set(user_question_df[user_question_df.num_question > num_question_per_user_thre]['user_id'])
print(len(popular_questions), len(common_users))
popular_questino_user_pair = simplified_df[simplified_df['question_id'].isin(popular_questions) & simplified_df['user_id'].isin(common_users)]
print(popular_questino_user_pair.shape)
popular_questino_user_pair.head(5)
```

```python
random_subset_b3 = popular_questino_user_pair # popular_questino_user_pair.sample(n=3000)
```

```python
random_subset_b3.to_csv(benchmark_3_data_file, sep=' ', header=False, index=False)
```
